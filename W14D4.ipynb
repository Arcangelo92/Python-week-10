```python
import os
import dotenv
from sqlalchemy import create_engine
import pandas as pd
dotenv.load_dotenv(override=True)
username = os.getenv("username")
password = os.getenv("password")
host = os.getenv("host")
dbname = os.getenv("dbname")
conn_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
db_engine = create_engine(conn_string)

query = "SELECT * FROM dimproduct"
df = pd.read_sql(query, db_engine)
db_engine.dispose ()
df.shape

#In media quanto pesano i prodotti? 
Peso_medio = round(df['Weight'].mean(),2)
print("Peso medio:", Peso_medio)

#• Quanto pesa il più leggero? 
Leggero = df['Weight'].min()
print("Peso più leggero:", Leggero)

#• Quanto pesa il più pesante? 
Leggero = df['Weight'].max()
print("Peso più leggero:", Leggero)

#• Quanti prodotti pesano più di 100 Kg? 
Leggero = df[df['Weight'] > 100][['EnglishProductName', 'Weight']]
print(Leggero.to_string(index=False))

#• Quanto costano in media i prodotti (colonna DealerPrice)? 
Costo_medio = df['ListPrice'].mean ()
print("il costo medio è: ", round(Costo_medio,2))

#• Se prendiamo un quarto di tutti i prodotti, in modo che siano i più costosi, quale sarà il loro range di prezzo?
quarto = df.sort_values(by='ListPrice', ascending=False)
totale = len(quarto)
top_quarto = quarto.head(int(totale / 4))
range_alto = top_quarto['ListPrice'].max()
range_basso = top_quarto['ListPrice'].min()
print("Prezzo tra:", range_alto, "e", range_basso)
```

```python
#• Qual è il prezzo medio per i prodotti di colore blu? 
lista_prezzi = df['ListPrice']
Blue = df[df['Color'] == 'Blue']
media_prezzi_blu = Blue['ListPrice'].mean()
print("Prezzo medio per i prodotti di colore Blue:", round(media_prezzi_blu, 2))

#• Qual è il prezzo medio per i prodotti di colore rosso o nero? 
lista_prezzi = df['ListPrice']
Red_balck = df[(df['Color'] == 'Red') | (df['Color'] == 'Black')]

media_prezzi_red_black = Red_balck['ListPrice'].mean()
print("Prezzo medio per i prodotti di colore Red o Black:", round(media_prezzi_red_black, 2))

#• Qual è il prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 Kg? 
specifiche = df[(df['Size'] == 42) & (df['Weight'] > 10)]
prezzo_max = specifiche['ListPrice'].max()
print("Il prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 Kg è:", prezzo_max)

#• Qual è il nome inglese e il costo di produzione (StandardCost) di tutti i prodotti di taglia 42, peso oltre i 10 Kg e colore argento? 
specifiche = df[(df['Size'] == 42) & (df['Weight'] > 10) & (df['Color'] == 'argento')]
nome_inglese = specifiche['EnglishDescription']
costo_di_produzione = specifiche['StandardCost']
print(nome_inglese, costo_di_produzione)

#• Esaminiamo la differenza tra lo StandardCost e il DealerPrice degli ultimi 20 elementi del dataset: 
#c'è un pattern? C'è qualche elemento che non lo segue?

ultimi_20 = df.tail(20)
ultimi_20['Differenza'] = ultimi_20['StandardCost'] - ultimi_20['DealerPrice']
print("Differenza tra StandardCost e DealerPrice degli ultimi 20 elementi del dataset:" , ultimi_20[['StandardCost', 'DealerPrice', 'Differenza']])
pattern = ultimi_20['Differenza'].value_counts()
print("\nConteggio delle differenze:")
print(pattern)
if len(pattern) == 1:
    print("\nTutte le differenze sono uguali, quindi c'è un pattern.")
else:
    print("\nCi sono differenze diverse, quindi ci potrebbe essere qualche elemento che non segue il pattern.")
```

```python
## DIABETES CSV

import pandas as pd
pd.read_csv('diabetes.csv')
DCSV = pd.read_csv('diabetes.csv')

a = DCSV.shape
b = DCSV.sample(5)  #Osserviamone le dimensioni e un'anteprima di cinque righe;
c = DCSV.dtypes #Prendiamoci un po' di tempo per dare un'occhiata ai metadati delle colonne;
d = DCSV.columns #Stampiamo dei descrittori statistici del dataset;
print (d)
media_bmi = DCSV['Body mass index (weight in kg/(height in m)^2)'].mean()
media_eta = DCSV['Age (years)'].max()
media_numero_gravidanze = DCSV['Number of times pregnant'].median()
print ('media_bmi :' ,media_bmi,'media_eta:' ,media_eta, 'media_numero_gravidanze: ', media_numero_gravidanze)

#Selezioniamo i dati relativi a diverse fasce di età: <20, 20-30, 30-40, 40-50, >50;
intervalli_eta = [0, 20, 30, 40, 50, float('inf')]  # [0, 20), [20, 30), [30, 40), [40, 50), [50, infinito)
etichette_eta = ['<20', '20-30', '30-40', '40-50', '>50']
DCSV['Fascia di età'] = pd.cut(DCSV['Age (years)'], bins=intervalli_eta, labels=etichette_eta)
dati_fasce_eta = DCSV.groupby('Fascia di età', observed=False).size()
print("Dati relativi a diverse fasce di età:" , dati_fasce_eta)

#Qual è la media della pressione sanguigna diastolica per le diverse fasce di età?
intervalli_eta = [0, 20, 30, 40, 50, float('inf')]  
etichette_eta = ['<20', '20-30', '30-40', '40-50', '>50']
DCSV['Fascia di età'] = pd.cut(DCSV['Age (years)'], bins=intervalli_eta, labels=etichette_eta)
media_pressione_per_fascia_eta = DCSV.groupby('Fascia di età', observed=False)['Diastolic blood pressure (mm Hg)'].mean().reset_index()
print("Media della pressione diastolica (mm Hg) per ogni fascia di età:")
print(media_pressione_per_fascia_eta)

#media di pressione per anno
DCSV_age = DCSV['Age (years)']
media_pressione_per__eta = DCSV.groupby(DCSV_age)['Diastolic blood pressure (mm Hg)'].mean()
print ("Età:", DCSV_age, "media_pressione: ", round(media_pressione_per__eta, 2))
```

```python
import pandas as pd
pd.read_csv('insurance.csv')
ICSV = pd.read_csv('insurance.csv')

#Visualizziamone le dimensioni, un'anteprima, e osserviamo i nomi di colonna;
a = ICSV.shape
b = ICSV.columns
print (a,b)

#Quali sono le medie di charges rispetto a region? Ci sono differenze significative?
medie_spese_per_regione = ICSV.groupby('region')['charges'].mean().reset_index()
print("\nMedie delle spese (charges) rispetto alla regione:")
print(round(medie_spese_per_regione,2))
#E rispetto a smoker? E a sex?
medie_smsx_per_regione = ICSV.groupby(['sex', 'smoker']).size().unstack()
print(round(medie_smsx_per_regione,2))

#Quali sono i descrittori statistici di bmi? Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?
desc_bmi = ICSV['bmi'].describe()
print(desc_bmi)
quartili_bmi = ICSV['bmi'].quantile([0.25, 0.5, 0.75])
min_media_max_charges_per_quartile_bmi = ICSV.groupby(pd.cut(ICSV['bmi'], quartili_bmi, include_lowest=True)).agg({'charges': ['min', 'mean', 'max']})
print(min_media_max_charges_per_quartile_bmi)
```
